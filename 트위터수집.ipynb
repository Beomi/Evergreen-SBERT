```python
from datetime import date

def get_today():
    return str(date.today())
```

```python
def search_keyword(q, max_results=1000):
    # 검색어 Query
    # GET https://api.twitter.com/1.1/search/tweets.json

    result = set()
    current_page = 1
    base_url = 'https://api.twitter.com/1.1/search/tweets.json'
    url = base_url + f'?q="{q}" lang:ko exclude:retweets'

    while (len(result) < max_results) and (current_page < 30):
        try:
            response = requests.get(
                url,
                params={
                    "count": "100",
                    "include_entities": "false",
                    'tweet_mode': 'extended',
                },
                headers={
                    "Authorization": "Bearer AAAAAAAAAAAAAAAAAAAAAI74QAEAAAAAEEWraMEnR6P%2F6UbrFkNuGN9t3Xs%3DTrFWb2a6uqjcnAZrf49Qxr6XCWtN2yAbPiqQk9Z2xK80lcKgRE",
                },
            )
            print('Response HTTP Status Code: {status_code}'.format(
                status_code=response.status_code))
            print(len(result), current_page)
            # print('Response HTTP Response Body: {content}'.format(
                # content=response.content))
            try:
                res = response.json()
                result.update([i['full_text'].split('https://t.co/')[0] for i in res['statuses'] if i['lang'] =='ko'])
            except KeyError as e:
                if 'statuses' in str(e):
                    print(res)
                    break
            current_page += 1
            try:
                url = base_url + res['search_metadata']['next_results']
            except KeyError as e:
                if 'next_results' in str(e):
                    print(res)
                    break
        except requests.exceptions.RequestException:
            print('HTTP Request failed')

    return result

```

```python
from pathlib import Path
```

```python
import json

def get_and_save(q):
    base_folder = Path(f'data/{get_today()}/')
    base_folder.mkdir(parents=True, exist_ok=True)
    base_folder.exists()
    json_path = base_folder/f'{q}.json'
    if json_path.exists():
        return json.load(open(json_path))
    data = search_keyword(q)
    json.dump(obj=list(data), fp=open(json_path, 'w'), ensure_ascii=False, indent=2)
    return data
```
